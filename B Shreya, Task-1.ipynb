# Import Modules

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
from sklearn import metrics 
```

# Load Dataset

```python
df = pd.read_excel(r"C:\Users\VBALA\Downloads\task1shreya.xlsx")
X_train = df["Hours"].values.reshape(-1, 1)
y_train = df["Scores"].values
```

# Fitting a Model

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

# Make Predictions

```python
y_pred = np.array([[9.2]])
output = model.predict(y_pred)
print("Predicted Score for given input 9.25 hours/day= ",output[0])
```

# Visualise the Model

```python
plt.scatter(X_train, y_train, color='blue')
plt.plot(X_train, model.predict(X_train), color='red', linewidth=2)
plt.scatter(y_pred, output, color='green')
plt.xlabel("Hours Studied Per Day")
plt.ylabel("Scores")
plt.title("Linear Regression Model to Predict Scores Given the Number of Hours Studied Per Day")
plt.show()
```

# Model Evaluation

```python
y_pred = model.predict(X_train)
mse = mean_squared_error(y_train, y_pred)
r2 = r2_score(y_train, y_pred)

print('Mean Absolute Error:', metrics.mean_absolute_error(y_train, y_pred)) 
# MAE gives the absolute error between actual and predicted value
# So, MAE of 4.9 includes the actual value can be +/- 4.9 to the predicted value
print(f"R-squared: {r2}") 
# R2 value close to 1 is considered good because it means that 
# a large portion of the variance in the target variable is captured by the model.
```
